```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import random

client_id = '313b0b0511374484a009f122c243767c'
client_secret = '7b76e974bd634cf5bbd3938ca98b0c06'

client_credentials_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

def get_track_features(track_ids):
    features = []
    for track_id in track_ids:
        try:
            track_info = sp.track(track_id)
            track_feature = sp.audio_features(track_id)[0]
            if track_feature:
                features.append({
                    'name': track_info['name'],
                    'artist': ', '.join([artist['name'] for artist in track_info['artists']]),
                    'features': track_feature
                })
        except Exception as e:
            print(f"Error fetching features for track ID: {track_id}, Error: {e}")
    return features

def generate_random_songs(num_songs=1000):
    try:

        playlist_id = '37i9dQZEVXbMDoHDwVN2tF'  
        playlist_tracks = sp.playlist_tracks(playlist_id)
        track_ids = [track['track']['id'] for track in playlist_tracks['items'] if track['track']['id'] is not None]


        selected_track_ids = random.sample(track_ids, min(num_songs, len(track_ids)))


        track_features = get_track_features(selected_track_ids)

        return track_features
    except Exception as e:
        print(f"Error: {e}")
        return None


random_songs_data = generate_random_songs(num_songs=1000)

if random_songs_data is not None:

    df = pd.DataFrame(random_songs_data)


    df.to_csv('random_songs_data.csv', index=False)
    print("Data exported to random_songs_data.csv")


    print(df.head())
else:
    print('Failed to fetch random songs')
```
